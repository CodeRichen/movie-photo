```python
#GIF去背
from PIL import Image
import os

# Load the GIF
gif_path = r"C:\Users\User\Downloads\w.gif"
gif = Image.open(gif_path)

# Directory to save frames
frames_dir = r"C:\Users\User\Downloads"
os.makedirs(frames_dir, exist_ok=True)
def remove_background(image_path, threshold=240):
    img = Image.open(image_path).convert("RGBA")
    datas = img.getdata()
    
    new_data = []
    for item in datas:
        # Change all white (also shades of whites)
        # pixels to transparent
        if item[0] > threshold and item[1] > threshold and item[2] > threshold:
            new_data.append((255, 255, 255, 0))
        else:
            new_data.append(item)
    
    img.putdata(new_data)
    return img
# Extract frames and save them
frame_paths = []
processed_frame_paths = []
try:
    while True:
        frame_number = gif.tell()
        frame_path = os.path.join(frames_dir, f"frame3_{frame_number}.png")
        gif.save(frame_path, format="PNG")
        img = remove_background(frame_path)
        processed_frame_path = frame_path.replace("/frames/", "/processed_frames/")
        processed_frame_paths.append(processed_frame_path)
        os.makedirs(os.path.dirname(processed_frame_path), exist_ok=True)
        img.save(processed_frame_path, "PNG")
        frame_paths.append(frame_path)
        gif.seek(frame_number + 1)
except EOFError:
    pass  # End of sequence

processed_frame_paths
```

```python
#使用API自動下載風景照片到資料夾中
import os
import requests
import ctypes

# 設定目標資料夾和 API 密鑰
IMAGE_DIR = "C:/Users/User/Desktop/landscape"
IMAGE_COUNT = 30  # 每次下載 30 張圖片
PIXABAY_API_KEY = "46066523-2648377b876c7e3d8198af60a"  # 在 Pixabay 註冊後獲取
EXISTING_IMAGES_FILE = os.path.join(IMAGE_DIR, 'downloaded_pixabay_images.txt')

SPI_SETDESKWALLPAPER = 20  # 設置桌布的 Windows API 常量

def load_existing_images():
    """讀取已下載的圖片URL清單"""
    if os.path.exists(EXISTING_IMAGES_FILE):
        with open(EXISTING_IMAGES_FILE, "r") as f:
            return set(line.strip() for line in f.readlines())
    return set()

def save_new_images(new_images):
    """保存新下載的圖片URL到紀錄檔"""
    with open(EXISTING_IMAGES_FILE, "a") as f:
        for img_url in new_images:
            f.write(f"{img_url}\n")

def set_lock_screen_wallpaper(image_path):
    """設置鎖定螢幕的壁紙"""
    ctypes.windll.user32.SystemParametersInfoW(SPI_SETDESKWALLPAPER, 0, image_path, 0)
    print(f"鎖定螢幕桌布已設置為: {image_path}")

def download_images_from_pixabay(image_dir, count):
    """從 Pixabay API 下載圖片"""
    url = f"https://pixabay.com/api/?key={PIXABAY_API_KEY}&image_type=photo&per_page={count}"
    response = requests.get(url)

    if response.status_code == 200:
        images = response.json()['hits']
        downloaded = 0
        new_images = []
        downloaded_images_paths = []

        existing_images = load_existing_images()

        for img in images:
            img_url = img['largeImageURL']  # 獲取高清圖片URL
            if img_url not in existing_images:
                img_response = requests.get(img_url, stream=True)

                if img_response.status_code == 200:
                    image_filename = os.path.join(image_dir, f"wallpaper_{downloaded + 1}.jpg")
                    with open(image_filename, "wb") as out_file:
                        out_file.write(img_response.content)

                    # print(f"成功下載圖片: {img_url}")
                    downloaded += 1
                    downloaded_images_paths.append(image_filename)
                    new_images.append(img_url)

                    if downloaded >= count:
                        break
                else:
                    print(f"下載圖片失敗: HTTP 狀態碼 {img_response.status_code}")
            else:
                print(f"已經下載過圖片: {img_url}")

        # 保存新下載的圖片URL到紀錄檔
        save_new_images(new_images)

        # 設置第2張圖片為鎖屏壁紙並刪除圖片
        if len(downloaded_images_paths) >= 2:
            lock_screen_image = downloaded_images_paths[1]
            # set_lock_screen_wallpaper(lock_screen_image)
            # os.remove(lock_screen_image)
            # print(f"已刪除: {lock_screen_image}")
        else:
            print("沒有足夠的圖片來設置鎖屏壁紙")
    else:
        print(f"API 請求失敗: HTTP 狀態碼 {response.status_code}")

def main():
    # 確保目標資料夾存在
    if not os.path.exists(IMAGE_DIR):
        os.makedirs(IMAGE_DIR)

    # 從 Pixabay 下載圖片
    download_images_from_pixabay(IMAGE_DIR, IMAGE_COUNT)

if __name__ == "__main__":
    main()
```

```python
import os
import requests
import time

API_KEY = '1XY58vfmVOdJ8vJGw3JO0BGLiiidmiGQ'
SEARCH_URL = 'https://wallhaven.cc/api/v1/search'
IMAGE_DIR = 'C:/Users/User/Desktop/wallpapers'
EXISTING_IMAGES_FILE = os.path.join(IMAGE_DIR, 'downloaded_wallhaven_images.txt')

def load_existing_images():
    """讀取已下載的圖片URL清單"""
    if os.path.exists(EXISTING_IMAGES_FILE):
        with open(EXISTING_IMAGES_FILE, "r") as f:
            return set(line.strip() for line in f.readlines())
    return set()

def save_new_images(new_images):
    """保存新下載的圖片URL到紀錄檔"""
    with open(EXISTING_IMAGES_FILE, "a") as f:
        for img_url in new_images:
            f.write(f"{img_url}\n")

def download_image(img_url, idx, retries=3):
    """下載圖片，失敗時重試"""
    for attempt in range(retries):
        try:
            img_response = requests.get(img_url, stream=True, timeout=10)
            if img_response.status_code == 200:
                image_filename = os.path.join(IMAGE_DIR, f"wallpaper_{idx + 1}.jpg")
                with open(image_filename, 'wb') as img_file:
                    img_file.write(img_response.content)
                print(f"圖片 {img_url} 下載成功")
                return image_filename  # 返回下載的圖片路徑
            else:
                print(f"下載失敗，狀態碼: {img_response.status_code}, 圖片: {img_url}")
        except requests.RequestException as e:
            print(f"下載錯誤: {e}，圖片: {img_url}，嘗試次數: {attempt + 1}/{retries}")
        time.sleep(2)  # 每次重試前等待 2 秒
    return None

def search_wallhaven(query, total_count=30):
    downloaded_images_paths = []
    new_images = []
    existing_images = load_existing_images()
    
    current_page = 1  # 當前頁數
    while len(downloaded_images_paths) < total_count:
        params = {
            'apikey': API_KEY,
            'q': query,
            'categories': 'anime',
            'purity': '100',  # 確保 SFW 圖片
            'sorting': 'random',
            'resolutions': '1920x1080,2560x1440',
            'page': current_page
        }

        response = requests.get(SEARCH_URL, params=params)
        
        if response.status_code == 200:
            data = response.json()
            for img in data['data']:
                img_url = img['path']
                if img_url in existing_images:
                    print(f"已經下載過圖片: {img_url}")
                    continue

                print(f"正在下載圖片: {img_url}")
                image_filename = download_image(img_url, len(downloaded_images_paths))
                if image_filename:
                    downloaded_images_paths.append(image_filename)
                    new_images.append(img_url)
                    existing_images.add(img_url)  # 更新已下載列表
                else:
                    print(f"下載失敗，跳過圖片: {img_url}")
                
                # 當下載的圖片數量達到目標時，跳出循環
                if len(downloaded_images_paths) >= total_count:
                    break
            current_page += 1  # 如果沒有達到下載數量，則檢索下一頁
        else:
            print("API 請求失敗", response.status_code)
            break

    # 保存新下載的圖片URL
    save_new_images(new_images)
    print(f"下載完成，共下載 {len(downloaded_images_paths)} 張圖片")

def main():
    # 確保目標資料夾存在
    if not os.path.exists(IMAGE_DIR):
        os.makedirs(IMAGE_DIR)

    # 從 Wallhaven 搜索並下載圖片
    search_wallhaven('anime', 30)

if __name__ == "__main__":
    main()
```
