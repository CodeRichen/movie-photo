```python
#GIF去背
from PIL import Image
import os

# Load the GIF
gif_path = r"C:\Users\User\Downloads\w.gif"
gif = Image.open(gif_path)

# Directory to save frames
frames_dir = r"C:\Users\User\Downloads"
os.makedirs(frames_dir, exist_ok=True)
def remove_background(image_path, threshold=240):
    img = Image.open(image_path).convert("RGBA")
    datas = img.getdata()
    
    new_data = []
    for item in datas:
        # Change all white (also shades of whites)
        # pixels to transparent
        if item[0] > threshold and item[1] > threshold and item[2] > threshold:
            new_data.append((255, 255, 255, 0))
        else:
            new_data.append(item)
    
    img.putdata(new_data)
    return img
# Extract frames and save them
frame_paths = []
processed_frame_paths = []
try:
    while True:
        frame_number = gif.tell()
        frame_path = os.path.join(frames_dir, f"frame3_{frame_number}.png")
        gif.save(frame_path, format="PNG")
        img = remove_background(frame_path)
        processed_frame_path = frame_path.replace("/frames/", "/processed_frames/")
        processed_frame_paths.append(processed_frame_path)
        os.makedirs(os.path.dirname(processed_frame_path), exist_ok=True)
        img.save(processed_frame_path, "PNG")
        frame_paths.append(frame_path)
        gif.seek(frame_number + 1)
except EOFError:
    pass  # End of sequence

processed_frame_paths
```

```python
# 下載30不重複初音的圖片到資料夾當中
import os
import time
import shutil
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from PIL import Image  # 用於進行後期處理

# 設定下載目標資料夾和 URL
IMAGE_DIR = "C:/Users/User/Desktop/wallpapers"
IMAGE_COUNT = 30  # 每次下載 30 張圖片
YANDEX_URL = "https://yandex.com/images/search?text=miku&isize=wallpaper&wp=wh16x10_2560x1600"
EXISTING_IMAGES_FILE = os.path.join(IMAGE_DIR, "downloaded_images.txt")  # 紀錄已下載圖片的URL

def load_existing_images():
    """讀取已下載的圖片URL清單"""
    if os.path.exists(EXISTING_IMAGES_FILE):
        with open(EXISTING_IMAGES_FILE, "r") as f:
            return set(line.strip() for line in f.readlines())
    return set()

def save_new_images(new_images):
    """保存新下載的圖片URL到紀錄檔"""
    with open(EXISTING_IMAGES_FILE, "a") as f:
        for img_url in new_images:
            f.write(f"{img_url}\n")

def enhance_image(image_path):
    """提升圖片畫質：這裡使用簡單的後處理來改善圖片品質"""
    with Image.open(image_path) as img:
        img = img.convert("RGB")
        img = img.resize((2560, 1600), Image.LANCZOS)  # 使用高質量的重採樣方式
        img.save(image_path, "JPEG", quality=95)  # 保存為 JPEG，質量設定為 95

def download_images(image_dir, count):
    """從 Yandex 下載圖片，避免重複下載已經下載過的圖片"""
    # 初始化 ChromeDriver
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # 隱藏瀏覽器窗口
    options.add_argument('--disable-gpu')
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    
    driver.get(YANDEX_URL)
    
    # 等待圖片加載
    time.sleep(5)

    # 找到所有圖片元素
    images = driver.find_elements(By.XPATH, "//img[@src]")  # 改用更通用的 XPATH 選擇器
    
    if not images:
        print("無法找到圖片，檢查網站結構")
        driver.quit()
        return

    existing_images = load_existing_images()
    downloaded = 0
    new_images = []
    
    for img in images:
        img_url = img.get_attribute("src")
        if img_url and img_url.startswith("//"):
            img_url = "https:" + img_url  # 處理相對路徑

        # 避免重複下載，並嘗試選擇大圖而非縮略圖
        if img_url and img_url not in existing_images:
            if "thumb" in img_url:  # 如果是縮略圖，嘗試找到大圖
                img_url = img_url.replace("thumb", "large")

            print(f"正在下載圖片: {img_url}")
            try:
                img_response = requests.get(img_url, stream=True)
                img_response.raw.decode_content = True
                
                # 儲存圖片
                image_filename = os.path.join(image_dir, f"wallpaper_{downloaded + 1}.jpg")
                with open(image_filename, "wb") as out_file:
                    shutil.copyfileobj(img_response.raw, out_file)

                # 壓縮並提升畫質
                enhance_image(image_filename)
                downloaded += 1
                new_images.append(img_url)

                if downloaded >= count:
                    break
            except Exception as e:
                print(f"下載圖片失敗: {e}")

    driver.quit()
    
    # 保存新下載的圖片URL到紀錄檔
    save_new_images(new_images)
    
    print(f"已下載 {downloaded} 張圖片")

def main():
    # 確保目標資料夾存在
    if not os.path.exists(IMAGE_DIR):
        os.makedirs(IMAGE_DIR)
    
    # 下載圖片
    download_images(IMAGE_DIR, IMAGE_COUNT)

if __name__ == "__main__":
    main()
```

```python
import os
import time
import shutil
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

# 定義網站對應的資料夾
WEBSITE_NAME = "pixiv"
ROOT_DIR = f"C:/Users/User/Desktop/{WEBSITE_NAME}_wallpapers"
IMAGE_COUNT = 10  # 每次下載 10 張圖片

PIXIV_URL = "https://www.pixiv.net/tags/%E5%8B%95%E6%BC%AB/artworks?s_mode=s_tag_full"

def create_website_folder(root_dir):
    """創建網站名稱的資料夾"""
    if not os.path.exists(root_dir):
        os.makedirs(root_dir)
    return root_dir

def login_pixiv(driver):
    """自動登入 Pixiv（如果需要）"""
    driver.get("https://accounts.pixiv.net/login")
    
    # 等待用戶手動輸入帳號密碼
    print("請手動登入 Pixiv...")
    time.sleep(30)  # 給用戶 30 秒時間手動登入，這個時間可以根據需要調整

def download_images(image_dir, count):
    """從 Pixiv 下載圖片"""
    # 初始化 ChromeDriver
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # 隱藏瀏覽器窗口
    options.add_argument('--disable-gpu')
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    
    # 開啟 Pixiv 登入頁面，等待手動登入
    login_pixiv(driver)

    # 登入後訪問圖片頁面
    driver.get(PIXIV_URL)
    
    # 等待圖片加載
    time.sleep(5)

    # 找到所有圖片元素
    images = driver.find_elements(By.XPATH, "//img[@srcset]")  # 使用 srcset 選擇器尋找高清圖片
    
    if not images:
        print("無法找到圖片，檢查網站結構")
        driver.quit()
        return

    downloaded = 0
    for img in images:
        img_url = img.get_attribute("srcset").split(",")[-1].split(" ")[0]  # 取高解析度的圖片URL
        
        if img_url:
            print(f"正在下載圖片: {img_url}")
            try:
                img_response = requests.get(img_url, stream=True)
                img_response.raw.decode_content = True
                
                # 儲存圖片
                image_filename = os.path.join(image_dir, f"anime_wallpaper_{downloaded + 1}.jpg")
                with open(image_filename, "wb") as out_file:
                    shutil.copyfileobj(img_response.raw, out_file)
                
                downloaded += 1

                if downloaded >= count:
                    break
            except Exception as e:
                print(f"下載圖片失敗: {e}")

    driver.quit()
    print(f"已下載 {downloaded} 張圖片到資料夾: {image_dir}")

def main():
    # 確保網站名稱資料夾存在
    folder_path = create_website_folder(ROOT_DIR)
    
    # 下載圖片到該資料夾
    download_images(folder_path, IMAGE_COUNT)

if __name__ == "__main__":
    main()
```
