```python
#GIF去背
from PIL import Image
import os

# Load the GIF
gif_path = r"C:\Users\User\Downloads\w.gif"
gif = Image.open(gif_path)

# Directory to save frames
frames_dir = r"C:\Users\User\Downloads"
os.makedirs(frames_dir, exist_ok=True)
def remove_background(image_path, threshold=240):
    img = Image.open(image_path).convert("RGBA")
    datas = img.getdata()
    
    new_data = []
    for item in datas:
        # Change all white (also shades of whites)
        # pixels to transparent
        if item[0] > threshold and item[1] > threshold and item[2] > threshold:
            new_data.append((255, 255, 255, 0))
        else:
            new_data.append(item)
    
    img.putdata(new_data)
    return img
# Extract frames and save them
frame_paths = []
processed_frame_paths = []
try:
    while True:
        frame_number = gif.tell()
        frame_path = os.path.join(frames_dir, f"frame3_{frame_number}.png")
        gif.save(frame_path, format="PNG")
        img = remove_background(frame_path)
        processed_frame_path = frame_path.replace("/frames/", "/processed_frames/")
        processed_frame_paths.append(processed_frame_path)
        os.makedirs(os.path.dirname(processed_frame_path), exist_ok=True)
        img.save(processed_frame_path, "PNG")
        frame_paths.append(frame_path)
        gif.seek(frame_number + 1)
except EOFError:
    pass  # End of sequence

processed_frame_paths
```

```python
import os
import time
import shutil
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from PIL import Image

# 設定下載目標資料夾和 URL
IMAGE_DIR = "C:/Users/User/Desktop/wallpapers"
IMAGE_COUNT = 30  # 每次下載 30 張圖片
YANDEX_URL = "https://yandex.com/images/search?text=miku&isize=wallpaper&wp=wh16x10_2560x1600"

def download_images(image_dir, count):
    """從 Yandex 下載圖片"""
    # 初始化 ChromeDriver
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # 隱藏瀏覽器窗口
    options.add_argument('--disable-gpu')
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    
    driver.get(YANDEX_URL)
    
    # 等待圖片加載
    time.sleep(5)

    # 找到所有圖片元素
    images = driver.find_elements(By.TAG_NAME, "img")
    
    if not images:
        print("無法找到圖片，檢查網站結構")
        driver.quit()
        return

    downloaded = 0
    for img in images:
        img_url = img.get_attribute("src")
        if img_url and img_url.startswith("//"):
            img_url = "https:" + img_url  # 處理相對路徑
        
        if img_url:
            print(f"正在下載圖片: {img_url}")
            try:
                img_response = requests.get(img_url, stream=True)
                img_response.raw.decode_content = True
                
                # 儲存圖片
                image_filename = os.path.join(image_dir, f"wallpaper_{downloaded + 1}.jpg")
                with open(image_filename, "wb") as out_file:
                    shutil.copyfileobj(img_response.raw, out_file)
                
                # 壓縮圖片為桌布大小
                compress_image(image_filename)
                downloaded += 1

                if downloaded >= count:
                    break
            except Exception as e:
                print(f"下載圖片失敗: {e}")

    driver.quit()
    print(f"已下載 {downloaded} 張圖片")

def compress_image(image_path):
    """壓縮圖片至桌布大小"""
    from PIL import Image
    with Image.open(image_path) as img:
        img = img.resize((2560, 1600), Image.Resampling.LANCZOS)  # 調整至目標大小
        img.save(image_path, optimize=True, quality=85)  # 壓縮圖片

def set_wallpaper(image_path):
    """設置單張圖片為桌布"""
    import ctypes
    SPI_SETDESKWALLPAPER = 20
    ctypes.windll.user32.SystemParametersInfoW(SPI_SETDESKWALLPAPER, 0, image_path, 3)

def change_wallpapers_periodically(image_dir, interval):
    """每隔一定時間更換一次桌布"""
    while True:
        # 下載新圖片
        download_images(image_dir, IMAGE_COUNT)
        
        # 確保資料夾內有圖片
        image_files = os.listdir(image_dir)
        if image_files:
            # 選擇一張圖片設置為桌布
            image_path = os.path.join(image_dir, image_files[0])
            set_wallpaper(image_path)
        else:
            print("資料夾內無圖片，無法設置桌布")

        # 每隔指定時間更換一次桌布
        print(f"等待 {interval} 秒後更換桌布...")
        time.sleep(interval)

def main():
    # 設置每隔 30 分鐘（1800 秒）更換一次桌布
    change_wallpapers_periodically(IMAGE_DIR, 1800)

if __name__ == "__main__":
    main()
```
